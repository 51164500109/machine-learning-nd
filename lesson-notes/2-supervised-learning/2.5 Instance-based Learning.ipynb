## KNN 

Given: 
- Training data D={x_i,y_i}
- Distance metric d(q,x) <- Represents domain knowledge
- Number of neighbours k <- Also represents
- Query point

Algorithm:
- $NN = \{ i: d(q,x_i) \text{ k smallest} \}$
    - If there are more than k that are closest, just take all of them. So take smallest number $\geq$ k.
Return:
- Classification: Vote of $y_i \in NN$, take the plurality (which one occurs the most). You can tiebreak e.g. by picking randomly or by picking the one that's more frequent in the entire dataset.
    - Could also do a weighted vote (weights depend on how far away you are). E.g. weight by 1/distance.
- Regression: Take the mean of the $y_i$s. Don't have to worry about a tie.

Simple algorithm but a lot left up to the designer.

### Running Time and Space

Given n sorted data points in R1 mapping to labels in R1.

1-NN query running time: binary search. Query space: constant because data storage accounted for in learning space.

KNN query running time : lgn + k, similar to merge sort. If k >= n/2 it dominates -> O(n), else log n dominates. Query space is constant because we can point in place. List is sorted so only need to point to first and last nearest neighbours.

Linear Regression
- Learning running time: involves inverting a matrix but this is scalar laned so we can just scan through the list to populate a constant-size matrix. n.
- Learning space: 1 (m and b)
- Query running time and space: 1

KNN learning is fast and querying is slow. With linear regression, learning is expensive and querying is cheap.
- If we query more than n times, NN is worse in terms of running time.
- Tradeoff: Want to balance the two.
- NN: Put off doing any work until you have to. **Lazy** learners vs linear regression **eager** learner.

### How KNN alg works

e.g. R2 -> R
- Distance metrics
    - Euclidean distance metric
    - Manhattan distance (l1?) (The distance between two points in a grid based on a strictly horizontal and/or vertical path (that is, along the grid lines), as opposed to the diagonal or "as the crow flies" distance.)

Different k and distance metrics can give completely different answers depending on the assumptions you make about your domain.

- KNN tends to work well.

### Preference biases of KNN
*Our belief about what makes a good hypothesis.*
- Locality -> Near points are similar
    - Further biases depending on distance function used
- Smoothness (Expecting functions to behave smoothly) -> Averaging (Think intermediate value theorem or something)
- ALl features matter equally (as opposed to $y = x_1^2 + x_2$.)
