# LSE daily data: Description and exploratory

#### Data Used
The data used is daily stock data for stocks on the London Stock Exchange (LSE). The date range for stock data varies depending on when the stock went public. The furthest date was **YEAR**. The most recent date in the dataset was 9 September 2016. The data was taken from Quandl.

All the data is in one CSV, with each row being one datapoint. Unless otherwise indicated, the column values are all floats. Each row includes:
* Stock symbol (string)
* Date (YYYY-MM-DD) 
* Open (given to 2 decimal places)
* High (to 2 d.p.)
* Low (to 2 d.p.)
* Close (to 2 d.p.)
* Volume (to 1 d.p.)
* Ex-Dividend (to 1 d.p.)
* Split Ratio (to 1 d.p.)
* Adjusted Open (to 6 d.p.)
* Adjusted High (to 6 d.p.)
* Adjusted Low (to 6 d.p.)
* Adjusted Close (to 6 d.p.)
* Adjusted Volume (to 1 d.p.)

That means we have 12 features for each stock on every trading day since the year when the stock was tradable.

# Feature Engineering
x-day running averages


````python
# N-day running averages
moving_average = 30

# 3-day, 7-day, 10-day, 14-day moving averages.
# Use a function

for i in range(moving_average, len(bp)):
    m_average = sum(bp.iloc[i-moving_average:i]['Adj. Close'])/moving_average
    bp.iloc[i].loc[:,'%s-day Moving Average' % str(moving_average)] = m_average
````

### Build training and test sets

````python
# Initialise variables
# Number of days prior that we consider
days = 7
# Number of train and test examples combined
periods = 9000

# Columns
columns = []
for j in range(1,days+1):
    columns.append('i-%s' % str(j))
columns.append('Adj. High')
columns.append('Adj. Low')
print(columns)

# Index
start_date = bp.iloc[days]["Date"]
print("Start date: ", start_date)
index = pd.date_range(start_date, periods=periods, freq='D')

# Create empty dataframes for features and prices
features = pd.DataFrame(index=index, columns=columns)
prices = pd.DataFrame(index=index, columns=["Target"])

# Prepare test and training sets
for i in range(periods):
    prices.iloc[i]['Target'] = bp.iloc[i+days]['Adj. Close']
    for j in range(days):
        features.iloc[i]['i-%s' % str(7-j)] = bp.iloc[i+j]['Adj. Close']
    features.iloc[i]['Adj. High'] = max(bp[i:i+days]['Adj. High'])
    features.iloc[i]['Adj. Low'] = min(bp[i:i+days]['Adj. Low'])
print(features.head())
print(prices.head())
````
